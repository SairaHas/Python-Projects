```python
#to drop those rows where the column cast has null values
df_with_cast=df.dropna(axis=0, subset=['cast'])
```

```python
df.info()
```

```python
cast_df=df_with_cast[['cast']]
```

```python
cast_df.head()
```

```python
#initialize an empty dictionary

cast_names_counter={}
```

```python
#this displayed the first four rows of the cast details
for index,row in cast_df.iterrows():
  movie_cast=row['cast']
  print(movie_cast)
  if index==5:
    break
```

```python
#now we need to split the cast names based on the comma separately,also there are white spaces which needs to be stripped
for index,row in cast_df.iterrows():
  movie_cast=row['cast']
  movie_cast_split=movie_cast.split(',')
  movie_cast_stripped=[name.strip().lower() for name in movie_cast_split] #also making every cast names into lowercases 
  print(movie_cast_stripped)
  if index==5:
    print()
    break
```

```python
#make the cast names lower cases and count the names of the cast using a counter
for index,row in cast_df.iterrows():
  movie_cast=row['cast']
  movie_cast_split=movie_cast.split(',')
  movie_cast_stripped=[name.strip().lower() for name in movie_cast_split]
  for name in movie_cast_stripped:
    cast_names_counter[name]=cast_names_counter.get(name,0)+1
 
```

```python
#the dictionary is sorted in the descending order calling out the name of the actors and the number of contents of each added in netflix
dict(sorted(cast_names_counter.items(),key=lambda item:item[1],reverse=True))
```
